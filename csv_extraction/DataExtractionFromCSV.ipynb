***Another way of writing python code to load CSV file data***

```python
import csv

def load_data(filename):
    myList= []
    with open(filename) as numbers:
        numbers_data = csv.reader(numbers, delimiter=',')
        next(numbers_data) #skip the header
        for row in numbers_data:
            myList.append(row)
        return myList
new_list = load_data('C:/Users/jyoth/Downloads/Amazon Sale Report.csv/Amazon Sale Report.csv')
for row in new_list[:10]:  # Print only the first 10 rows. 
    #print(row)
    display(row)
```

***Saving Extracted Data to a CSV File***

After extracting data from CSV, you can save it into a new CSV file using Pandas in Python.

```python
import pandas as pd
sales_csv =('C:/Users/jyoth/Downloads/Amazon Sale Report.csv/Amazon Sale Report.csv')
sales_df = pd.read_csv(sales_csv)
#print(sales_df)
display(sales_df)
display(sales_df.head(50))

# Save Data to Another CSV File

output_csv = 'C:/Users/jyoth/Downloads/Amazon_Sale_Report_Copy.csv'
sales_df.to_csv(output_csv, index=False, encoding="utf-8")

print(f" Data saved to {output_csv}")
```

**sales_df.to_csv(output_csv, index=False, encoding="utf-8")**

index=False: 

This option prevents Pandas from including the DataFrame's index (row labels) as an extra column in the saved CSV file, resulting in a cleaner output without unnecessary data.

encoding="utf-8": 

Specifies the character encoding for the CSV file, ensuring that all characters, including special and non-ASCII characters, are correctly written and can be properly read by other systems.

***Inserting extracted data into MySQL***

You can also store the extracted data in a SQL database

```python
import pandas as pd
import pyodbc
from sqlalchemy import create_engine

# Load Data from CSV
sales_csv = 'C:/Users/jyoth/Downloads/Amazon Sale Report.csv/Amazon Sale Report.csv'
sales_df = pd.read_csv(sales_csv, encoding="utf-8", low_memory=False)

# SQL Server Connection Details 

db_server = 'JYOTHI\SQLEXPRESS'  # SQL Server instance
db_name = 'dataExtraction'         # SQL Server database name

# Connection string for SQL Server using pyodbc

conn_str = f'DRIVER={{SQL Server}};SERVER={db_server};DATABASE={db_name};Trusted_Connection=yes;'

try:
    # Create SQLAlchemy engine using pyodbc for SQL Server
    
    engine = create_engine(f'mssql+pyodbc:///?odbc_connect={conn_str}')
    
    # Connect to SQL Server and insert the DataFrame
    
    with engine.connect() as conn:
        sales_df.to_sql('sales', conn, if_exists='replace', index=False)
        print("Data inserted into SQL Server database successfully!")
except Exception as e:
    print(f"Error inserting data into SQL Server: {e}")
```

**sales_df = pd.read_csv(sales_csv, encoding="utf-8", low_memory=False)**

-Reads the CSV file into a Pandas DataFrame.

-encoding="utf-8" ensures proper character encoding.

-low_memory=False avoids data type issues when handling large files.

**conn_str = f'DRIVER={{SQL Server}};SERVER={db_server};DATABASE={db_name};Trusted_Connection=yes;'**

-Specifies the database driver and server name.

-Uses Trusted_Connection=yes to authenticate using Windows credentials.

**engine = create_engine(f'mssql+pyodbc:///?odbc_connect={conn_str}')**

-Establishes a connection to SQL Server using SQLAlchemy and pyodbc.

-Converts the connection string into a format SQLAlchemy understands.

**with engine.connect() as conn:**
    
   **sales_df.to_sql('sales', conn, if_exists='replace', index=False)**

-Connects to SQL Server and writes the DataFrame (sales_df) to a table named "sales".

-if_exists='replace' ensures the table is replaced if it already exists.

-index=False prevents Pandas from adding an extra index column in SQL Server.
